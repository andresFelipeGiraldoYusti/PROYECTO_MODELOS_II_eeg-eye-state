```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Configuraci√≥n de visualizaci√≥n
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

print("="*60)
print("AN√ÅLISIS EXPLORATORIO DE DATOS PARA ML")
print("="*60)

# ============================================
# 2. VISI√ìN GENERAL DEL DATASET
# ============================================
def analisis_general(df):
    print("\nüìä INFORMACI√ìN GENERAL")
    print("-"*60)
    print(f"Dimensiones: {df.shape[0]} filas x {df.shape[1]} columnas")
    print(f"Memoria utilizada: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    print(f"\nPrimeras filas:")
    display(df.head())

    print("\nüìã TIPOS DE DATOS")
    print("-"*60)
    display(pd.DataFrame({
        'Tipo': df.dtypes,
        'No Nulos': df.count(),
        'Nulos': df.isnull().sum(),
        '% Nulos': (df.isnull().sum() / len(df) * 100).round(2)
    }))

    print(f"\nüî¢ Duplicados: {df.duplicated().sum()} ({df.duplicated().sum()/len(df)*100:.2f}%)")

analisis_general(df)

# ============================================
# 3. AN√ÅLISIS DE VALORES FALTANTES
# ============================================
def analizar_valores_faltantes(df):
    print("\n\n‚ùå AN√ÅLISIS DE VALORES FALTANTES")
    print("-"*60)

    missing = df.isnull().sum()
    missing_pct = (missing / len(df)) * 100
    missing_df = pd.DataFrame({
        'Columna': missing.index,
        'Valores_Faltantes': missing.values,
        'Porcentaje': missing_pct.values
    }).sort_values('Porcentaje', ascending=False)

    missing_df = missing_df[missing_df['Valores_Faltantes'] > 0]

    if len(missing_df) > 0:
        display(missing_df)

        # Visualizaci√≥n
        plt.figure(figsize=(12, 6))
        sns.heatmap(df.isnull(), cbar=True, yticklabels=False, cmap='viridis')
        plt.title('Patr√≥n de Valores Faltantes')
        plt.tight_layout()
        plt.show()
    else:
        print("‚úÖ No hay valores faltantes")

analizar_valores_faltantes(df)

# ============================================
# 4. AN√ÅLISIS DE VARIABLES NUM√âRICAS
# ============================================
def analizar_numericas(df):
    print("\n\nüî¢ AN√ÅLISIS DE VARIABLES NUM√âRICAS")
    print("-"*60)

    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

    if len(num_cols) == 0:
        print("No hay variables num√©ricas")
        return

    # Estad√≠sticas descriptivas
    print("\nüìà Estad√≠sticas Descriptivas:")
    display(df[num_cols].describe().round(2))

    # Distribuciones
    n_cols = min(len(num_cols), 3)
    n_rows = (len(num_cols) - 1) // n_cols + 1

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4*n_rows))
    axes = axes.flatten() if len(num_cols) > 1 else [axes]

    for i, col in enumerate(num_cols):
        axes[i].hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
        axes[i].set_title(f'Distribuci√≥n: {col}')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Frecuencia')

        # A√±adir l√≠nea de media
        mean_val = df[col].mean()
        axes[i].axvline(mean_val, color='red', linestyle='--',
                       label=f'Media: {mean_val:.2f}')
        axes[i].legend()

    # Ocultar subplots vac√≠os
    for i in range(len(num_cols), len(axes)):
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()

    # Boxplots para detectar outliers
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4*n_rows))
    axes = axes.flatten() if len(num_cols) > 1 else [axes]

    for i, col in enumerate(num_cols):
        axes[i].boxplot(df[col].dropna(), vert=True)
        axes[i].set_title(f'Boxplot: {col}')
        axes[i].set_ylabel(col)

    for i in range(len(num_cols), len(axes)):
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()

    # Detecci√≥n de outliers (IQR)
    print("\nüéØ DETECCI√ìN DE OUTLIERS (M√©todo IQR):")
    for col in num_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower) | (df[col] > upper)][col]
        print(f"{col}: {len(outliers)} outliers ({len(outliers)/len(df)*100:.2f}%)")

analizar_numericas(df)

# ============================================
# 5. MATRIZ DE CORRELACI√ìN
# ============================================
def analizar_correlaciones(df):
    print("\n\nüîó MATRIZ DE CORRELACI√ìN")
    print("-"*60)

    num_cols = df.select_dtypes(include=[np.number]).columns

    if len(num_cols) < 2:
        print("Se necesitan al menos 2 variables num√©ricas")
        return

    corr_matrix = df[num_cols].corr()

    # Heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
                center=0, square=True, linewidths=1)
    plt.title('Matriz de Correlaci√≥n')
    plt.tight_layout()
    plt.show()

    # Correlaciones altas (> 0.7 o < -0.7)
    print("\n‚ö†Ô∏è Correlaciones Altas (|r| > 0.7):")
    high_corr = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) > 0.7:
                high_corr.append({
                    'Var1': corr_matrix.columns[i],
                    'Var2': corr_matrix.columns[j],
                    'Correlaci√≥n': corr_matrix.iloc[i, j]
                })

    if high_corr:
        display(pd.DataFrame(high_corr))
    else:
        print("No hay correlaciones altas entre variables")

analizar_correlaciones(df)

# ============================================
# 6. AN√ÅLISIS DE VARIABLES CATEG√ìRICAS
# ============================================
def analizar_categoricas(df):
    print("\n\nüìä AN√ÅLISIS DE VARIABLES CATEG√ìRICAS")
    print("-"*60)

    cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

    if len(cat_cols) == 0:
        print("No hay variables categ√≥ricas")
        return

    for col in cat_cols:
        print(f"\nüìå Variable: {col}")
        print(f"Valores √∫nicos: {df[col].nunique()}")

        value_counts = df[col].value_counts()
        value_pct = df[col].value_counts(normalize=True) * 100

        result_df = pd.DataFrame({
            'Frecuencia': value_counts,
            'Porcentaje': value_pct.round(2)
        })

        display(result_df.head(10))

        # Gr√°fico de barras (top 10)
        plt.figure(figsize=(10, 5))
        value_counts.head(10).plot(kind='bar', edgecolor='black')
        plt.title(f'Top 10 Valores: {col}')
        plt.xlabel(col)
        plt.ylabel('Frecuencia')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

analizar_categoricas(df)

# ============================================
# 7. AN√ÅLISIS DE BALANCE DE CLASES
# ============================================
def analizar_variable_objetivo(df, target_col):
    """
    Analiza la variable objetivo para problemas de clasificaci√≥n
    """
    if target_col not in df.columns:
        print(f"\n‚ö†Ô∏è La columna '{target_col}' no existe")
        return

    print(f"\n\nüéØ AN√ÅLISIS DE VARIABLE OBJETIVO: {target_col}")
    print("-"*60)

    value_counts = df[target_col].value_counts()
    value_pct = df[target_col].value_counts(normalize=True) * 100

    balance_df = pd.DataFrame({
        'Clase': value_counts.index,
        'Frecuencia': value_counts.values,
        'Porcentaje': value_pct.values.round(2)
    })

    display(balance_df)

    # Visualizaci√≥n
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Gr√°fico de barras
    axes[0].bar(balance_df['Clase'].astype(str), balance_df['Frecuencia'],
                edgecolor='black')
    axes[0].set_title('Distribuci√≥n de Clases')
    axes[0].set_xlabel('Clase')
    axes[0].set_ylabel('Frecuencia')

    # Gr√°fico de pastel
    axes[1].pie(balance_df['Frecuencia'], labels=balance_df['Clase'],
                autopct='%1.1f%%', startangle=90)
    axes[1].set_title('Proporci√≥n de Clases')

    plt.tight_layout()
    plt.show()

    # Detectar desbalanceo
    min_class = balance_df['Porcentaje'].min()
    max_class = balance_df['Porcentaje'].max()
    ratio = max_class / min_class

    if ratio > 2:
        print(f"\n‚ö†Ô∏è Dataset DESBALANCEADO (ratio: {ratio:.2f}:1)")
        print("üí° Considera t√©cnicas como SMOTE, class_weight o undersampling")
    else:
        print(f"\n‚úÖ Dataset relativamente balanceado (ratio: {ratio:.2f}:1)")

# Ejecutar si tienes variable objetivo
# analizar_variable_objetivo(df, 'tu_variable_objetivo')

# ============================================
# 8. REPORTE FINAL Y RECOMENDACIONES
# ============================================
def generar_reporte_final(df):
    print("\n\n" + "="*60)
    print("üìã REPORTE FINAL Y RECOMENDACIONES")
    print("="*60)

    num_cols = df.select_dtypes(include=[np.number]).columns
    cat_cols = df.select_dtypes(include=['object', 'category']).columns

    print("\n‚úÖ ACCIONES RECOMENDADAS:")
    print("\n1. VALORES FALTANTES:")
    missing = df.isnull().sum()
    if missing.sum() > 0:
        print("   - Imputar valores faltantes (media/mediana/moda)")
        print("   - Considerar eliminar columnas con >50% de nulos")
    else:
        print("   ‚úì No hay valores faltantes")

    print("\n2. VARIABLES NUM√âRICAS:")
    print("   - Normalizar/Estandarizar (StandardScaler, MinMaxScaler)")
    print("   - Tratar outliers si es necesario")
    print("   - Aplicar transformaciones (log, sqrt) si hay asimetr√≠a")

    print("\n3. VARIABLES CATEG√ìRICAS:")
    if len(cat_cols) > 0:
        print("   - Encoding: OneHotEncoder, LabelEncoder, o Target Encoding")
        print("   - Agrupar categor√≠as raras (frecuencia < 5%)")

    print("\n4. FEATURE ENGINEERING:")
    print("   - Crear variables de interacci√≥n")
    print("   - Binning de variables continuas")
    print("   - Extracci√≥n de caracter√≠sticas temporales (si aplica)")

    print("\n5. REDUCCI√ìN DE DIMENSIONALIDAD:")
    if len(num_cols) > 10:
        print("   - Considerar PCA o SelectKBest")
        print("   - Eliminar variables altamente correlacionadas")

    print("\n6. VALIDACI√ìN:")
    print("   - Train/Test Split (70-30 o 80-20)")
    print("   - Cross-Validation (k-fold)")
    print("   - Estratificaci√≥n si hay desbalanceo")

generar_reporte_final(df)

print("\n" + "="*60)
print("‚úÖ AN√ÅLISIS COMPLETADO")
print("="*60)
```
