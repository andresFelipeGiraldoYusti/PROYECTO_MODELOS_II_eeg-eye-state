# Leer base de datos

```python
# === 1. Montar Google Drive ===
from google.colab import drive
drive.mount('/content/drive')

# === 2. Instalar librerías necesarias ===
!pip install -q liac-arff pandas

# === 3. Importar librerías ===
import arff
import pandas as pd
import os # Import the os module

# === 4. Definir la ruta del archivo ARFF ===
ruta = "/content/drive/MyDrive/Practica_MODELOS_II/data/phplE7q6h.arff"

# === 5. Verificar la existencia del archivo ===
if not os.path.exists(ruta):
    print(f"❌ Error: El archivo no se encontró en la ruta: {ruta}")
else:
    # === 6. Cargar el archivo ARFF ===
    with open(ruta, "r", encoding="utf-8") as f:
        arff_data = arff.load(f)

    # Extraer nombres de columnas
    columnas = [attr[0] for attr in arff_data['attributes']]

    # Crear DataFrame con pandas
    df = pd.DataFrame(arff_data['data'], columns=columnas)

    # === 7. Mostrar información básica ===
    print("✅ Archivo cargado correctamente")
    print("Dimensiones:", df.shape)
    print("Columnas:", columnas)
```

# Distribución de cada feacture

```python
# Crear figura con subplots
fig, axes = plt.subplots(3, 5, figsize=(20, 12))  # 3 filas, 5 columnas
axes = axes.flatten()  # Convertir a array 1D para fácil iteración

# Columnas V a graficar
columnas_v = [f'V{i}' for i in range(1, 15)]  # V1 a V14

for i, columna in enumerate(columnas_v):
    # Calcular límites para cada columna
    limite_inferior = df[columna].quantile(0.01)
    limite_superior = df[columna].quantile(0.99)

    # Filtrar datos
    df_filtrado = df[(df[columna] >= limite_inferior) & (df[columna] <= limite_superior)]

    # Crear boxplot en el subplot correspondiente
    sns.boxplot(data=df_filtrado, x=columna, ax=axes[i])
    axes[i].set_title(f'{columna}\n(P1-99: {limite_inferior:.1f} - {limite_superior:.1f})', fontsize=10)
    axes[i].tick_params(axis='x', rotation=45)

# Eliminar subplots vacíos si los hay
for i in range(len(columnas_v), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

# Mostrar estadísticas para cada columna
print("=== ESTADÍSTICAS DE FILTRADO ===")
for columna in columnas_v:
    limite_inferior = df[columna].quantile(0.01)
    limite_superior = df[columna].quantile(0.99)
    df_filtrado = df[(df[columna] >= limite_inferior) & (df[columna] <= limite_superior)]

    print(f"{columna}:")
    print(f"  • Datos mostrados: {len(df_filtrado)}/{len(df)} ({len(df_filtrado)/len(df)*100:.1f}%)")
    print(f"  • Rango: {limite_inferior:.2f} - {limite_superior:.2f}")
    print()
```

# Correlación entre variables

```python
# Calcular matriz de correlación
correlation_matrix = df.corr()

# Mostrar correlaciones con la variable objetivo Class
correlaciones_class = correlation_matrix['Class'].sort_values(ascending=False)
print("=== CORRELACIONES CON CLASS ===")
print(correlaciones_class)
```

```python
# Crear tabla ordenada de correlaciones con Class
correlaciones_df = pd.DataFrame({
    'Variable': correlaciones_class.index,
    'Correlacion': correlaciones_class.values,
    'Abs_Correlacion': np.abs(correlaciones_class.values)
}).sort_values('Abs_Correlacion', ascending=False)

print("=== CORRELACIONES ORDENADAS POR IMPORTANCIA ===")
print(correlaciones_df.to_string(index=False))

# Separar correlaciones positivas y negativas
print("\n=== CORRELACIONES POSITIVAS CON CLASS ===")
positivas = correlaciones_df[correlaciones_df['Correlacion'] > 0]
print(positivas.to_string(index=False))

print("\n=== CORRELACIONES NEGATIVAS CON CLASS ===")
negativas = correlaciones_df[correlaciones_df['Correlacion'] < 0]
print(negativas.to_string(index=False))
```

```python
# Heatmap de toda la matriz de correlación
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix,
            annot=True,
            cmap='RdBu_r',
            center=0,
            fmt='.2f',
            square=True,
            cbar_kws={'shrink': 0.8})
plt.title('Matriz de Correlación Completa', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

```python
# Crear matriz de correlación solo con Class
corr_with_class = correlation_matrix[['Class']].sort_values('Class', ascending=False)

plt.figure(figsize=(8, 12))
sns.heatmap(corr_with_class,
            annot=True,
            cmap='RdBu_r',
            center=0,
            fmt='.3f',
            cbar_kws={'shrink': 0.8})
plt.title('Correlaciones con Variable Class', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de barras de correlaciones con Class
plt.figure(figsize=(12, 8))
correlaciones_class_sin_self = correlaciones_class.drop('Class')  # Quitar autocorrelación
colors = ['red' if x < 0 else 'blue' for x in correlaciones_class_sin_self]

plt.barh(correlaciones_class_sin_self.index, correlaciones_class_sin_self.values, color=colors)
plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
plt.xlabel('Correlación con Class')
plt.title('Correlaciones de Variables con Class')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Top 10 variables más correlacionadas (positiva y negativamente)
top_correlaciones = correlaciones_class_sin_self.iloc[np.argsort(np.abs(correlaciones_class_sin_self))[-10:]]

plt.figure(figsize=(10, 6))
colors = ['red' if x < 0 else 'blue' for x in top_correlaciones]
plt.barh(top_correlaciones.index, top_correlaciones.values, color=colors)
plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
plt.xlabel('Correlación con Class')
plt.title('Top 10 Variables Más Correlacionadas con Class')
plt.grid(axis='x', alpha=0.3)

# Añadir valores en las barras
for i, (index, value) in enumerate(top_correlaciones.items()):
    plt.text(value, i, f' {value:.3f}',
             va='center',
             ha='left' if value > 0 else 'right',
             fontweight='bold')

plt.tight_layout()
plt.show()

print("=== TOP 10 CORRELACIONES MÁS FUERTES ===")
for var, corr in top_correlaciones.items():
    print(f"{var}: {corr:.3f}")
```

```python
# Clasificar correlaciones por fuerza
def clasificar_correlacion(valor):
    abs_val = abs(valor)
    if abs_val >= 0.7: return "Muy Fuerte"
    elif abs_val >= 0.5: return "Fuerte"
    elif abs_val >= 0.3: return "Moderada"
    elif abs_val >= 0.1: return "Débil"
    else: return "Muy Débil"

correlaciones_clasificadas = correlaciones_class_sin_self.apply(clasificar_correlacion)

# Crear resumen por categorías
resumen_correlaciones = pd.DataFrame({
    'Variable': correlaciones_class_sin_self.index,
    'Correlacion': correlaciones_class_sin_self.values,
    'Categoria': correlaciones_clasificadas.values
})

print("=== RESUMEN POR CATEGORÍAS DE CORRELACIÓN ===")
categoria_counts = resumen_correlaciones['Categoria'].value_counts()
print(categoria_counts)

print("\n=== VARIABLES POR CATEGORÍA ===")
for categoria in ['Muy Fuerte', 'Fuerte', 'Moderada', 'Débil', 'Muy Débil']:
    vars_categoria = resumen_correlaciones[resumen_correlaciones['Categoria'] == categoria]
    if not vars_categoria.empty:
        print(f"\n{categoria}:")
        for _, row in vars_categoria.iterrows():
            print(f"  {row['Variable']}: {row['Correlacion']:.3f}")
```

```python
# Seleccionar las 5 variables más correlacionadas con Class
top_5_vars = correlaciones_class_sin_self.nlargest(5).index.tolist()
top_5_vars.append('Class')  # Añadir la variable objetivo

# Crear pairplot (puede ser pesado computacionalmente)
try:
    sns.pairplot(df[top_5_vars], hue='Class', diag_kind='hist', palette='viridis')
    plt.suptitle('Pairplot de Variables Más Correlacionadas con Class', y=1.02)
    plt.show()
except Exception as e:
    print(f"Pairplot muy pesado, mostrando scatter plots individuales...")

    # Scatter plots individuales
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    axes = axes.flatten()

    for i, var in enumerate(top_5_vars[:4]):  # Primeras 4 variables
        sns.scatterplot(data=df, x=var, y='Class', ax=axes[i], alpha=0.6)
        axes[i].set_title(f'{var} vs Class (corr: {correlation_matrix.loc[var, "Class"]:.3f})')

    plt.tight_layout()
    plt.show()
```

```python
# Filtrar solo columnas V y Class
columnas_v = [f'V{i}' for i in range(1, 15)] + ['Class']
corr_v_class = df[columnas_v].corr()['Class'].sort_values(ascending=False)

print("=== CORRELACIONES DE COLUMNAS V CON CLASS ===")
for var, corr in corr_v_class.items():
    if var != 'Class':
        print(f"{var}: {corr:.3f}")

# Gráfico específico para columnas V
plt.figure(figsize=(10, 6))
corr_v_sin_class = corr_v_class.drop('Class')
colors_v = ['red' if x < 0 else 'blue' for x in corr_v_sin_class]

plt.bar(corr_v_sin_class.index, corr_v_sin_class.values, color=colors_v)
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.xticks(rotation=45)
plt.xlabel('Variables V')
plt.ylabel('Correlación con Class')
plt.title('Correlaciones de Columnas V con Class')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
